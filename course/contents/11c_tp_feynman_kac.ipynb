## Problème

On veut calculer $u(x, t)$ solution de :

$$\partial_t u + a(x)\,\partial_x u + \tfrac{1}{2} b^2(x)\,\partial_{xx} u = 0, \qquad u(x, T) = g(x)$$

Par Feynman-Kac : $u(x,t) = \mathbb{E}[g(X_T) \mid X_t = x]$ avec $dX_s = a(X_s)\,ds + b(X_s)\,dW_s$.
